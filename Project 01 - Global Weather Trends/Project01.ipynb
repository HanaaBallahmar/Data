```python
# Importing the important Libraries
import numpy as np
import pandas as pd   # To load data into the notebook
import matplotlib.pyplot as plt #To create line chart
import matplotlib.ticker as ticker
%matplotlib nbagg

# Importing the extracted Data Set
data = pd.read_csv("results.csv")

# Create dataframe
df = pd.DataFrame(data)
```

```python
# Calculate the mean of global average tempture 
mean_global = df['global_avg_temp'].mean()

# Detect missing value in global average tempture and filling it with mean
df['global_avg_temp'].fillna(mean_global, inplace = True)
        
```

```python
# Calculate the mean of local average tempture
mean_local = df['local_avg_temp'].mean()

# Detect missing value in local average tempture and assign value to it
df['local_avg_temp'].fillna(mean_local, inplace = True)   

df['GMA'] = df['global_avg_temp'].rolling(15).mean()
df['LMA'] = df['local_avg_temp'].rolling(15).mean()


df
```

```python
f, ax = plt.subplots()
year = ['1828','1843','1858','1873','1888','1903','1918','1933','1948','1963','1978','1993','2008', '2023']
ax.set_xticklabels(year)
plt.title('Temperature in Makkak vs global temperature', fontsize=16)
ax.plot(df['GMA'], label='Moving Average Global (15 Years)', color='orange')
ax.plot(df['LMA'], label='Moving Average Makkah (15 Years)', color='blue')

plt.xlabel ('Years',fontsize=14)
plt.ylabel ('Temperature (°C)', fontsize=14)
plt.ylim(0,30) 
#plt.xlim(1840,2020,15)

plt.legend(loc='upper left',bbox_to_anchor=(1.10, 0.8), shadow=True, ncol=1, fontsize=12)
plt.grid(True)
plt.show()
```
